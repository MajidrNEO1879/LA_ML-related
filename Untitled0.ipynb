```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../content/sample_data/Zomato-data-.csv")
df.head(100)
```

```python
df['rate'] = df['rate'].str.replace(' ', '/', regex=False)
df['rate'] = df['rate'].str.split('/').str[0]
df['rate'] = df['rate'].astype(float)
df
```

```python
yes_count = (df['online_order'] == 'Yes').sum()
no_count = (df['online_order'] == 'No').sum()
plt.bar(['Yes', 'No'], [yes_count, no_count], color=['green', 'red'])
plt.xlabel("Online Order")
plt.ylabel("Count")
plt.title("Order Histogram")
plt.show()
```

```python
din_num = 0;
buffet_num = 0;
Cafes_num = 0;
others = 0;
for i in df['listed_in(type)']:
  if i == "Dining":
    din_num += 1;
  elif i == "Buffet":
    buffet_num += 1;
  elif i == "Cafes":
    Cafes_num += 1;
  else:
    others +=1;
din_num
```

```python
plt.bar(['Buffet', 'Others', 'Cafe', 'Dining'], [buffet_num,others, Cafes_num,din_num])
plt.show()
```

```python
rating_counts = df['approx_cost(for two people)'].value_counts().sort_index()
plt.bar(rating_counts.index, rating_counts.values, color='blue')
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Rating Distribution")
plt.show()
```
