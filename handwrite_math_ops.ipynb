```python
path = untar_data(URLs.MNIST_SAMPLE)
Path.BASE_PATH = path
path.ls()
```

```python
zero_img_path = (path/'train'/'3').ls()[18]
Image.open(zero_img_path)
```

```python
zero_img_sample_tensor = tensor(Image.open( zero_img_path))
zero_img_sample_tensor
```

```python
df = pd.DataFrame(zero_img_sample_tensor[:15])
df.style.set_properties(**{'font-size': '6pt'}).background_gradient('Blues')
```

```python
def label_func(x):
    return x.parent.name
dls = ImageDataLoaders.from_path_func(
    path=path,
    fnames=get_image_files(path),
    valid_pct=0.2,
    seed=42,
    label_func=label_func,
    item_tfms=Resize(224)
)
```

```python
learn = cnn_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(1)
```

```python
uploader = widgets.FileUpload()
uploader
```

```python
img = PILImage.create(uploader.data[0])
prediction,_,probs = learn.predict(img)
print(f'Prediction: {prediction}')
print(probs)
```
